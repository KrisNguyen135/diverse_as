```python
import numpy as np
import pandas as pd
```

```python
subset_data_ids = np.load(f'../../data/vast/model_50000_1/ids.npy')

microblogs = pd.read_pickle('../../data/vast/microblogs_dataset.pkl')
microblogs = microblogs.iloc[subset_data_ids, :]

microblogs
```

```python
keywords = [
    'sore', 'throat', 'fever', 'fatigu', 'cough', 'short', 'breath', 'chill',
    'sick', 'pain', 'diarrhea', 'stomach', 'sweat', 'pneumonia', 'flu', 'ach',
    'nausea', 'vomit',  'nauseou', 'declin', 'health', 'headach', 'nose', 'runni'
]
```

```python
# [
#     'sore', 'fever', 'fatigu',
#     'sick', 'pain', 'sweat', 'ach',
#      declin', 'health', 'headach',
# ]

respiratory_kws = [
    "throat", "cough", "breadth", "pneumonia", "flu", "nose", "runni"
]

stomach_flu_kws = [
    "chill", "diarrhea", "stomach", "nausea", "vomit", "nauseou"
]

other_kws = [
    "sore", "fever", "fatigu", "sick", "pain", "sweat", "ach", "declin",
    "health", "headach"
]


def label_row(row):
    for word in row["porter_stems"]:
        if word in respiratory_kws:
            return 2
        elif word in stomach_flu_kws:
            return 3
        elif word in other_kws:
            return 4
    
    return 1


microblogs["label"] = microblogs.apply(lambda row: label_row(row), axis=1)
microblogs
```
