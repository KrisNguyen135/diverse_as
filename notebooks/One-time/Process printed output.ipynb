```python
import numpy as np
import os
```

```python
path = '../../notes/bjob_output'
file = 'run.1398258'

with open(os.path.join(path, file), 'r') as f:
    lines = f.readlines()
    
lines
```

```python
line = lines[22]

start = line.index('(')
end = line.index(',')

query = int(line[(start + 1):end])
query
```

```python
train_ind = []

for line in lines:
    if not line.startswith('Exp '):
        continue
    
    start = line.index('(')
    try:
        end = line.index(',')
        train_ind.append(int(line[(start + 1):end]))
    except ValueError:
        assert 'Error' in line
    
train_ind
```

```python
train_ind = np.array(train_ind).reshape(-1, 1)
np.savetxt(f'train_ind_{file}', train_ind, fmt='%d')
```
