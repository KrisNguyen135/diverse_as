```python
import numpy as np
import os
```

```python
path = '../../notes'
file = 'run.1402319'

with open(os.path.join(path, file), 'r') as f:
    lines = f.readlines()
    
lines
```

```python
train_ind = []
labels = []

for line in lines:
    if not line.startswith('Exp '):
        continue
    
    # chosen ind
    start = line.index('(')
    end = line.index(',')
    train_ind.append(int(line[(start + 1):end]))
    
    # chosen label
    start = end
    end = line.index(')')
    labels.append(int(line[(start + 1):end]))
        
init_ind = int(lines[10].strip())
init_label = int(lines[12].strip())

train_ind = [init_ind] + train_ind
labels = [init_label] + labels
```

```python
train_ind = np.array(train_ind).reshape(-1, 1)
labels = np.array(labels).reshape(-1, 1)

np.savetxt('ens jensen greedy__ind__2.csv', train_ind, fmt='%d')
np.savetxt('ens jensen greedy__labels__2.csv', labels, fmt='%d')
```
