```python
import numpy as np
import os
import re
```

```python
path = "../../data/printed_output/das_vast"
file = "vast.1765132"

with open(os.path.join(path, file), "r") as f:
    lines = f.readlines()
    
lines
```

```python
train_ind = []
labels = []

for i, line in enumerate(lines):
    if not line.startswith("Exp "):
        continue
    
    # chosen ind
    start = line.index("(")
    end = line.index(",")
    train_ind.append(int(line[(start + 1):end]))
    
    # chosen label
    start = end
    end = line.index(")")
    labels.append(int(line[(start + 1):end]))
    
init_ind = int(lines[35].strip())
init_label = int(lines[37].strip())

train_ind = [init_ind] + train_ind
labels = [init_label] + labels
```

```python
train_ind = np.array(train_ind).reshape(-1, 1)
labels = np.array(labels).reshape(-1, 1)
```

```python
verify_flag = True
verify_path = "../../data/results/vast/3/ens jensen greedy_500"

exp = int(lines[13].strip())

try:
    ind_file = os.path.join(
        verify_path,
        f"ens jensen greedy_500__ind__{exp}.csv",
    )
        
    verify_ind = np.loadtxt(ind_file)
except:
    print("ind file not found")
    
    verify_flag = False
    
try:
    label_file = os.path.join(
        verify_path,
        f"ens jensen greedy_500__labels__{exp}.csv"
    )
        
    verify_labels = np.loadtxt(label_file)
except:
    print("label file not found")
    
    verify_flag = False
    
    
if verify_flag:
    print(np.isclose(train_ind.flatten(), verify_ind).all())
    print(np.isclose(labels.flatten(), verify_labels).all())
```

```python
if not verify_flag:
    np.savetxt(
        os.path.join(
            verify_path,
            f"ens jensen greedy_500__ind__{exp}.csv"
        ),
        train_ind,
        fmt="%d"
    )

    np.savetxt(
        os.path.join(
            verify_path,
            f"ens jensen greedy_500__labels__{exp}.csv"
        ),
        labels,
        fmt="%d"
    )
```
