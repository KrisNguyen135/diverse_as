```python
import numpy as np
import os
```

```python
path = '../../notes'
file = 'run.1402318'

with open(os.path.join(path, file), 'r') as f:
    lines = f.readlines()
    
lines
```

```python
train_ind = []

for line in lines:
    if not line.startswith('Exp '):
        continue
    
    start = line.index('(')
    try:
        end = line.index(',')
        train_ind.append(int(line[(start + 1):end]))
    except ValueError:
        assert 'Error' in line
        
init_ind = int(lines[10].strip())
train_ind = [init_ind] + train_ind
train_ind
```

```python
train_ind = np.array(train_ind).reshape(-1, 1)
np.savetxt(f'train_ind_{file}', train_ind, fmt='%d')
```
