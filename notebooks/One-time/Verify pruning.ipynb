```python
import numpy as np
import os
```

```python
policy = 'ens jensen greedy'

folder1 = 'ens jensen greedy_pruning'
name1 = 'queries'
folder2 = 'ens jensen greedy'
name2 = 'ind'

n_exps = 20
n_classes = 6
budget = 100

basepath = '../../data/results/square'

for i in range(n_exps):
    queries1 = np.loadtxt(
        os.path.join(basepath, folder1, f'{policy}__{name1}__{i + 1}.csv'),
        dtype=int
    ).flatten() - 1
    queries1 = queries1[-budget:]
    
    queries2 = np.loadtxt(
        os.path.join(basepath, folder2, f'{policy}__{name2}__{i + 1}.csv'),
        dtype=int
    ).flatten() - 1
    queries2 = queries2[-budget:]
    
    print(queries1)
    print(queries2)
    
    assert np.isclose(queries1, queries2).all()
```
