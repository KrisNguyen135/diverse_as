```python
import numpy as np
import os
import re
```

```python
path = '../../data/printed_output'
file = 'run.1402933'

nn_file = "nn.1400347"
num_negatives = 100_000
group_size = 4

# Read in the size of the search space in each problem
with open(os.path.join(path, nn_file), "r") as f:
    lines = f.readlines()
    
num_positives = []
for line in lines:
    if line.startswith("processing fingerprint gpidaph3 ..."):
        break
    
    if not line.startswith("	100000 inactives, "):
        continue
    
    num_positives.append(int(line.split(" ")[2]))
    
assert len(num_positives) == 120 // group_size

ns = np.array(num_positives) + num_negatives
```

```python
train_ind = []
labels = []
pruned = []

for i, line in enumerate(lines):
    if not line.startswith('Exp '):
        continue
    
    # chosen ind
    start = line.index('(')
    end = line.index(',')
    train_ind.append(int(line[(start + 1):end]))
    
    # chosen label
    start = end
    end = line.index(')')
    labels.append(int(line[(start + 1):end]))
    
    # pruning
    try:
        pruned.append(list(map(int, re.split("\s+", lines[i + 1].strip()))))
    except ValueError:
        print(line[i + 1])
        
init_ind = int(lines[35].strip())
init_label = int(lines[37].strip())

train_ind = [init_ind] + train_ind
labels = [init_label] + labels

pruned = np.array(pruned)
```
