```python
import torch
import torchvision

import numpy as np

import matplotlib.pyplot as plt

plt.style.use("bmh")
plt.rcParams["figure.figsize"] = (8, 6)
```

```python
train_loader = torchvision.datasets.FashionMNIST("./data", train=True, download=True)
test_loader = torchvision.datasets.FashionMNIST("./data", train=False, download=True)
```

```python
train_loader.data.shape, test_loader.data.shape
```

```python
np.save(
    "./data/labels.npy", 
    np.array(
        train_loader.targets.detach().tolist() + test_loader.targets.detach().tolist()
    )
)
```

```python
data = np.vstack(
    [train_loader.data.detach().numpy(), test_loader.data.detach().numpy()]
)

data.shape
```

```python
import umap
```

```python
embedding = umap.UMAP(n_components=10).fit_transform(data.reshape(data.shape[0], -1))
embedding.shape
```

```python
np.save("./data/embedding_10.npy", embedding)
```

```python
plt.scatter(
    embedding[:, 0], 
    embedding[:, 1], 
    c=train_loader.targets.detach().tolist() + test_loader.targets.detach().tolist(),
    alpha=0.3,
    s=1,
);
```

```python
plt.scatter(
    embedding[:, 2], 
    embedding[:, 3], 
    c=train_loader.targets.detach().tolist() + test_loader.targets.detach().tolist(),
    alpha=0.3,
    s=1,
);
```
