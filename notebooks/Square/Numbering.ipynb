```python
import numpy as np
from scipy.io import loadmat

import matplotlib.pyplot as plt
import seaborn as sns
import vapeplot; vapeplot.set_palette('vaporwave')
```

```python
data = loadmat('../../data/square_small/square_small_nearest_neighbors.mat')
nns = data['nearest_neighbors'] - 1
sims = data['similarities']
xs = data['x']
labels = data['labels']
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
cmap = plt.get_cmap('PuBu', labels.max())
c = ax.scatter(xs[:, 0], xs[:, 1], c=labels, cmap=cmap, edgecolors='grey', s=100)

for i in range(xs.shape[0]):
    ax.text(xs[i, 0], xs[i, 1], str(i + 1))

ax.set_aspect('equal', 'box')
plt.colorbar(c);
```
