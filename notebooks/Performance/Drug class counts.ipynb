```python
import numpy as np

import os

from statsmodels.stats.api import DescrStatsW
from scipy.stats import ttest_rel, sem

import matplotlib.pyplot as plt
import seaborn as sns

# %matplotlib ipympl
```

```python
def labels_to_counts(labels, time_axis=0):
    onehot = np.zeros((labels.size, n_classes), dtype=int)
    onehot[np.arange(labels.size), labels] = 1
    return onehot.cumsum(axis=time_axis)


data = 'ecfp'
policies = [
    'ens jensen greedy',
    # 'greedy',
    'round robin greedy'
]
linestyles = [
    '-',
    '--',
    # 'dotted'
]
assert len(policies) == len(linestyles)

n_groups = 1
n_exps = 5
n_classes = 5
init_size = 1
budget = 500

basepath = os.path.join('../../data/results', data)


class_counts = -1 * np.ones(
    (len(policies), n_groups, n_exps, init_size + budget, n_classes), 
    dtype=int
)

for p_ind, policy in enumerate(policies):
    for group_ind in range(n_groups):
        path = os.path.join(basepath, f'ecfp{group_ind + 1}', policy)
        for i in range(n_exps):
            labels = np.loadtxt(
                os.path.join(path, f'{policy}__labels__{i + 1}.csv'), 
                dtype=int
            ).flatten() - 1

            class_counts[p_ind, group_ind, i, :] = labels_to_counts(labels)
        
assert np.all(class_counts != -1)
```

```python
plt.figure(figsize=(10, 8))

for i, (policy, style) in enumerate(zip(policies, linestyles)):
    for c in range(1, n_classes):
        plt.plot(
            class_counts[i, :, :, :, c].mean(axis=(0, 1)), 
            linestyle=style, c=f'C{c - 1}', label=f'class {c} by {policy}'
        )
        
plt.legend();
```

```python
fig, ax = plt.subplots(1, len(policies), figsize=(8 * len(policies), 5), sharey=True)

for i, (policy, style) in enumerate(zip(policies, linestyles)):
    for c in range(1, n_classes):
        ax[i].plot(
            class_counts[i, :, :, :, c].mean(axis=(0, 1)),
            linestyle=style, c=f'C{c - 1}', label=f'class {c + 1}'
        )
        
        ax[i].legend()
        ax[i].set_title(policy)
        
plt.show()
```
